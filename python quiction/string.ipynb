```python
# palindrome number
class Palindrome:
    def is_palindrome(self, s):
        return s == s[::10]

p = Palindrome()
print(p.is_palindrome("racecar"))  
```

```python
# simple calculater


class Calculater():
    pass
    def add(self):
        self.num1 = int(input("enter the number: "))
        self.num2 = int(input("enter the number: "))
         
        print("addition")
        print(self.num1 + self.num2)

    def subtract(self):
        self.num1 = int(input("enter the number: "))
        self.num2 = int(input("enter the number: "))
        
        print("subtraction")
        print(self.num1 - self.num2)

    def multiply(self):
        self.num1 = int(input("enter the number: "))
        self.num2 = int(input("enter the number: "))

        print("multiplaction")
        print(self.num1 * self.num2)

    def divide(self):
        self.num1 = int(input("enter the number: "))
        self.num2 = int(input("enter the number: "))
        
        print("division")
        print(self.num1 / self.num2)

        

c = Calculater()
while True:
    print("-----------calculater-----------")
    print("1. addition")
    print("2. subtraction")
    print("3. multiplay")
    print("4. divided")
    print("5. exit")

    option = int(input("enter your option: "))

    match option:
        case 1:
            c.add()
        case 2:
            c.subtract()
        case 3:
            c.multiply()
        case 4:
            c.divide()
        case 5:
            print("thank you")
            break


```

```python

#  read csv file
import pandas as pd
sale = pd.read_csv("sale.csv")
print(sale)
```

```python
# deleat duplicate rows

print(sale.drop_duplicates())
```

```python
#  top 10  rows

print(sale.head(10))
```

```python
# last 10 rows
print(sale.tail(10))
```

```python

# all columns name and type
print(sale.columns)
```

```python

# information
print(sale.info())
```

```python
#  discribe
print(sale.describe())
```

```python
# all region count(seprated by region)

region = sale["Region"].value_counts()
print(region)
total_region = sale["Region"].value_counts().sum()
print("total region", total_region)
```

```python
#  all region wise sale data

reagion_sale = sale.groupby("Region")["Sales"].sum()
print(reagion_sale)
```

```python

# region wise sales report
import matplotlib.pyplot as plt

plt.bar(reagion_sale.index, reagion_sale.values)
plt.title("All Region Sale Report",color = "red")
plt.xlabel("region")
plt.ylabel("sales")
plt.savefig("all_region_sale_report.png")
plt.show()
```

```python
#  year sale 

year_sale = sale.groupby("Year")["Sales"].sum()
print(year_sale)
```

```python
# year wise sales report

plt.bar(year_sale.index, year_sale.values)
plt.title("Year Sale Report", color = "red")
plt.xlabel("years")
plt.ylabel("sale")
plt.savefig("year_sale_report.png")
plt.show()
```

```python
# month sale 

month_sale = sale.groupby("Month")["Sales"].sum()
print(month_sale)
```

```python
# month wise sales report

plt.bar(month_sale.index, month_sale.values)
plt.title("Month Wise Sales Report", color = "red")
plt.xlabel("month")
plt.ylabel("sale")
plt.savefig("month_sale_report.png")
plt.show()
```

```python
# sales summery for all product

laptop = sale.pivot_table(values="Sales", index="Month", columns="Product", aggfunc="sum")
print(laptop)
```

```python
#  laptop sale report
print("laptop sales report")
laptop_sales = sale[sale["Product"] == "Laptops"].groupby("Month")["Sales"].sum()
print(laptop_sales)
```

```python
# Accessories sale report
print("Accessories Sale Report")
accessories_sale = sale[sale["Product"] == "Accessories"].groupby("Month")["Sales"].sum()
print(accessories_sale)
```

```python
#  Phone sale report
print("Phone Sale Report")
phone_sale = sale[sale["Product"]=="Phones"].groupby("Month")["Sales"].sum()
print(phone_sale)
```

```python
#  smartwatchess sale report
print("Smart Watchess Sale Report")
smartwatchess_sale = sale[sale["Product"] == "Smartwatches"].groupby("Month")["Sales"].sum()
print(smartwatchess_sale)
```

```python
#   TV sale report
print("TV Sale Report")
tv_sale = sale[sale["Product"] == "TVs"].groupby("Month")["Sales"].sum()
print(tv_sale)
```

```python
#  tablate sale report
print("tablet sales report")
tablets_sale = sale[sale["Product"] == "Tablets"].groupby("Month")["Sales"].sum()
print(tablets_sale)
```

```python
# visulize all product data one by one

plt.suptitle("Item Sales Report (Monthly Analysis)", fontsize=20, fontweight="bold")

plt.figure(figsize=(18, 12))
# bar chart - laptop sale per month
plt.subplot(3,2,1)
plt.bar(laptop_sales.index,laptop_sales.values,color = "red")
plt.title("Laptop sale per month")
plt.xlabel("Month")
plt.ylabel("Sale")

# bar chart - phone sale per month
plt.subplot(3,2,2)
plt.bar(phone_sale.index,phone_sale.values, color = "green")
plt.title("Phone sale per month")
plt.xlabel("Month")
plt.ylabel("Sale")

# bar chart - tablets sale per month
plt.subplot(3,2,3)
plt.bar(tablets_sale.index,tablets_sale.values,color = "blue")
plt.title("Tablets sale per month")
plt.xlabel("Month")
plt.ylabel("Sales")

# bar chart - smartwatchess sale per month
plt.subplot(3,2,4)
plt.bar(smartwatchess_sale.index,smartwatchess_sale.values,color = "Brown")
plt.title("Smartwatchess sale per month")
plt.xlabel("Month")
plt.ylabel("Sale")

# bar chart - Accessories sales per month
plt.subplot(3,2,5)
plt.bar(accessories_sale.index,accessories_sale.values,color = "BlueViolet")
plt.title("Accessories sales per month")
plt.xlabel("Month")
plt.ylabel("sale")

# bar chart - tv's sale per month
plt.subplot(3,2,6)
plt.bar(tv_sale.index,tv_sale.values,color = "Chocolate")
plt.title("TV's sale per month")
plt.xlabel("Month")
plt.ylabel("Sale")

plt.tight_layout()
plt.savefig("item_sale_report.pdf")
plt.show()
```
