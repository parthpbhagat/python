```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("sale.csv")
print(df)
```

```python
print(df.columns)
```

```python
reagen = df["Region"].value_counts()
print(reagen)
```

```python
reagen_sale = df.groupby("Region")["Sales"].sum()
print(reagen_sale)
```

```python
year_sale = df.groupby("Year")["Sales"].sum()
print(year_sale)
```

```python

# total sale per month
month_sale = df.groupby("Month")["Sales"].sum()
print(month_sale)
```

```python

# total profit for each product per month
laptop = df.groupby(["Product", "Month"])["Profit"].sum()
print(laptop)
```

```python

# table summery for all product
laptop = df.pivot_table(values="Sales", index="Month", columns="Product", aggfunc="sum")
print(laptop)
```

```python

# laptop sale per month
laptop_sales = df[df["Product"] == "Laptops"].groupby("Month")["Sales"].sum()
print(laptop_sales)
```

```python
# phone sale per month
phone_sale = df[df["Product"] == "Phones"].groupby("Month")["Sales"].sum()
print(phone_sale)
```

```python
# tablets sales per month
tablets_sale = df[df["Product"] == "Tablets"].groupby("Month")["Sales"].sum()
print(tablets_sale)
```

```python
# Smartwatches sales per month
smartwatches_sale = df[df["Product"] == "Smartwatches"].groupby("Month")["Sales"].sum()
print(smartwatches_sale)
```

```python
# tv's sale per month
tv_sale = df[df["Product"] == "TVs"].groupby("Month")["Sales"].sum()
print(tv_sale)
```

```python
# Accessories sales per month
Accessories_sale = df[df["Product"] == "Accessories"].groupby("Month")["Sales"].sum()
print(Accessories_sale)
```

```python



# item's vise sale report

plt.suptitle("Item Sales Report (Monthly Analysis)", fontsize=20, fontweight="bold")

plt.figure(figsize=(18, 12))
# bar chart - laptop sale per month
plt.subplot(3,2,1)
plt.bar(laptop_sales.index,laptop_sales.values,color = "red")
plt.title("Laptop sale per month")
plt.xlabel("Month")
plt.ylabel("Sale")

# bar chart - phone sale per month
plt.subplot(3,2,2)
plt.bar(phone_sale.index,phone_sale.values, color = "green")
plt.title("plone sale per month")
plt.xlabel("month")
plt.ylabel("sale")

# bar chart - tablets sale per month
plt.subplot(3,2,3)
plt.bar(tablets_sale.index,tablets_sale.values,color = "blue")
plt.title("tablets sale per month")
plt.xlabel("Month")
plt.ylabel("sales")

# bar chart - smartwatchess sale per month
plt.subplot(3,2,4)
plt.bar(smartwatches_sale.index,smartwatches_sale.values,color = "Brown")
plt.title("smartwatchess sale per month")
plt.xlabel("Month")
plt.ylabel("Sale")

# bar chart - Accessories sales per month
plt.subplot(3,2,5)
plt.bar(Accessories_sale.index,Accessories_sale.values,color = "BlueViolet")
plt.title("Accessories sales per month")
plt.xlabel("month")
plt.ylabel("sale")

# bar chart - tv's sale per month
plt.subplot(3,2,6)
plt.bar(tv_sale.index,tv_sale.values,color = "Chocolate")
plt.title("tv's sale per month")
plt.xlabel("month")
plt.ylabel("sale")

plt.tight_layout()
plt.savefig("item_sale_report.pdf")
plt.show()
```

```python
# total sale for eatch product per month
laptop = df.groupby(["Product", "Month"])["Sales"].sum()
print(laptop)
```

```python

# total profit for all product 
profit = df.groupby("Product")["Profit"].sum()
print(profit)
```

```python

#   bar chart - profit contribuction by product
plt.bar(profit.index,profit.values,color = "pink")
plt.title("profit contribuction by product")
plt.xlabel("product")
plt.ylabel("profit")
plt.tight_layout
plt.show()
```

```python

# profit per year
profit_year = df.groupby("Year")["Profit"].sum()
print(profit_year)
```

```python

#  bar chart - profit contribuction per year
plt.bar(profit_year.index,profit_year.values,color = "#40ff00")
plt.title("Profit contribuction Per year")
plt.xlabel("year")
plt.ylabel("Profit")
plt.show()
```

```python
#  all types of chart

plt.figure(figsize=(18, 12))
# bar plot
plt.subplot(3,2,1)
plt.bar(reagen_sale.index,reagen_sale.values,color = ["red","green","blue","purple"])
plt.title("Total Sale By Region")
plt.xlabel("Region")
plt.ylabel("Value")
plt.legend()

#line plot
plt.subplot(3,2,2)
plt.plot(year_sale.index, year_sale.values,marker = ">")
plt.title("Total Year sale")
plt.xlabel("year")
plt.ylabel("value")
plt.grid()

# bar chart per month
plt.subplot(3,2,3)
plt.bar(month_sale.index, month_sale.values,color = "orange")
plt.xlabel("month")
plt.ylabel("sales")
plt.title("Total Sales per Month ")

# pie chart
plt.subplot(3,2,4)
plt.pie(profit.values,labels=profit.index,autopct="%1.1f%%")
plt.title("Prifit Contribuction By Product")

# line chart - profit contribuction per year
plt.subplot(3,2,5)
plt.plot(profit_year.index,profit_year.values, linestyle="--",marker = ">",color = "green")
plt.title("Profit Contribuction Per Year")
plt.ylabel("Profit")
plt.xlabel("Year")
plt.grid()

# scatter plot - profit contribuction per year
plt.subplot(3,2,6)
plt.scatter(df["Sales"],df["Profit"],c=df["Year"], cmap="viridis", s=40, alpha=0.6)
plt.colorbar(label = "year")
plt.title("Sales Vs Profit")
plt.xlabel("sales")
plt.ylabel("profit")


plt.tight_layout()
plt.savefig("sale_chart.pdf")
plt.show()
```
