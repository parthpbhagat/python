```python

x = df["MemberID"].head(10)
y1 = df["BMI"].head(10)
y2 = df["Height_cm"].head(10)
y3 = df["Weight_kg"].head(10)
y4 = df["Heart_Rate"].head(10)
plt.plot(x,y1,label = "BMI", color= "red", marker = "o")
plt.plot(x,y2,label = "Hight", color= "g", marker = "^")
plt.plot(x,y3, label = "weight", color= "blue", marker = ">")
plt.legend( title = "summery")
plt.title("line chart")
plt.grid()
plt.show()
```

```python
x = df["MemberID"].head(10)
y1 = df["BMI"].head(10)
y2 = df["Height_cm"].head(10)
y3 = df["Weight_kg"].head(10)
y4 = df["Heart_Rate"].head(10)
plt.figure(figsize=(10, 8))
# BMI chart
plt.subplot(2, 2, 1)
plt.title("BMI chart")
plt.bar(x,y1,color = "red")
plt.ylabel("BMI")
plt.xlabel("Member id")


plt.subplot(2, 2, 2)
plt.title("Hight chart")
plt.bar(x,y2, color = "green")
plt.ylabel("hight")
plt.xlabel("Member id")


plt.subplot(2, 2, 3)
plt.title("Weight")
plt.bar(x,y3, color = "blue")
plt.ylabel("weight")
plt.xlabel("Member id")

plt.subplot(2, 2, 4)
plt.title("Heart_Rate")
plt.bar(x,y4,color = "purple")
plt.ylabel("hart rate")
plt.xlabel("Member id")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("fitness_1000.csv")

# Bar chart for Gender frequency
df["Gender"].value_counts().plot(kind="bar")
plt.title("Bar Chart - Gender Frequency")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.savefig("bar_chart_gender.png")  # saves image in same folder
plt.show()
```

```python


gender = (df["Gender"]).value_counts().plot(kind="barh",color = "orange")
print(gender)
plt.title("Gender counter")
plt.xlabel("gender")
plt.ylabel("count")
plt.legend()
plt.tight_layout()
plt.show()


```

```python

gender = (df["Gender"]).value_counts().plot(kind="bar")
plt.title("gender counter")
plt.xlabel("gender")
plt.ylabel("count")
plt.legend()
plt.show()
```

```python
gender_1 = df["Gender"].value_counts()
print(gender_1)
plt.bar(gender_1.index,gender_1.values,color = ["red","blue"])
plt.title("gender distributer")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend( "Gender")
plt.savefig("Gender.png")
plt.tight_layout()
plt.show()
```
